Алгоритмы шифрования
Шифр Цезаря
Симметричный алгоритм, в котором каждая буква в тексте заменяется буквой, находящейся на определенном количестве позиций левее или правее нее в алфавите.

Ключ: Целое число, указывающее на сколько позиций сдвигать буквы.
Пример ключа: 3
Шифр Плейфера
Блочный шифр, использующий матрицу 5x5 (или 6x6 для расширенного набора символов) для замены пар букв в тексте.

Ключ: Ключевое слово для создания матрицы.
Пример ключа: секрет
Шифр XOR
Каждый символ текста подвергается побитовой операции XOR с соответствующим символом ключа.

Ключ: Строка любого размера.
Пример ключа: password
Шифр RSA
Асимметричный алгоритм шифрования, использующий пару ключей: публичный для шифрования и приватный для расшифровки.

Ключи генерируются автоматически.
Приватный ключ необходимо сохранить для расшифровки.
Шифр Виженера
Использует повторяющееся ключевое слово для шифрования текста посредством сдвига каждой буквы на значение буквы ключа.

Ключ: Ключевое слово.
Пример ключа: ключ
Важные замечания
Кодировка: Убедитесь, что ваша консоль или терминал поддерживает кодировку UTF-8 для корректного отображения русских символов.
Размер сообщения в RSA: При использовании RSA максимальная длина сообщения ограничена размером ключа. По умолчанию используется ключ размером 2048 бит.

## rsa: Библиотека для реализации RSA-шифрования.

```
# -*- coding: utf-8 -*-

import base64
import rsa

# Определяем русский алфавит
alphabet = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
alphabet_lower = alphabet.lower()

letter_to_index = dict(zip(alphabet, range(len(alphabet))))
index_to_letter = dict(zip(range(len(alphabet)), alphabet))
letter_to_index_lower = dict(zip(alphabet_lower, range(len(alphabet))))
index_to_letter_lower = dict(zip(range(len(alphabet)), alphabet_lower))

# Шифр Цезаря
def caesar_encrypt(text, key):
    result = ''
    for char in text:
        if char.upper() in letter_to_index:
            idx = letter_to_index[char.upper()]
            new_idx = (idx + key) % len(alphabet)
            if char.isupper():
                result += index_to_letter[new_idx]
            else:
                result += index_to_letter_lower[new_idx]
        else:
            result += char
    return result

def caesar_decrypt(text, key):
    result = ''
    for char in text:
        if char.upper() in letter_to_index:
            idx = letter_to_index[char.upper()]
            new_idx = (idx - key) % len(alphabet)
            if char.isupper():
                result += index_to_letter[new_idx]
            else:
                result += index_to_letter_lower[new_idx]
        else:
            result += char
    return result

# Шифр Виженера
def vigenere_encrypt(text, keyword):
    result = ''
    keyword_repeat = ''
    keyword = keyword.upper()
    for i in range(len(text)):
        if text[i].upper() in letter_to_index:
            keyword_repeat += keyword[i % len(keyword)]
        else:
            keyword_repeat += text[i]
    for i in range(len(text)):
        char = text[i]
        key_char = keyword_repeat[i]
        if char.upper() in letter_to_index:
            idx = letter_to_index[char.upper()]
            key_idx = letter_to_index[key_char]
            new_idx = (idx + key_idx) % len(alphabet)
            if char.isupper():
                result += index_to_letter[new_idx]
            else:
                result += index_to_letter_lower[new_idx]
        else:
            result += char
    return result

def vigenere_decrypt(text, keyword):
    result = ''
    keyword_repeat = ''
    keyword = keyword.upper()
    for i in range(len(text)):
        if text[i].upper() in letter_to_index:
            keyword_repeat += keyword[i % len(keyword)]
        else:
            keyword_repeat += text[i]
    for i in range(len(text)):
        char = text[i]
        key_char = keyword_repeat[i]
        if char.upper() in letter_to_index:
            idx = letter_to_index[char.upper()]
            key_idx = letter_to_index[key_char]
            new_idx = (idx - key_idx) % len(alphabet)
            if char.isupper():
                result += index_to_letter[new_idx]
            else:
                result += index_to_letter_lower[new_idx]
        else:
            result += char
    return result

# Шифр XOR
def xor_encrypt(text, key):
    key_length = len(key)
    encrypted_bytes = bytearray()
    for i in range(len(text)):
        encrypted_char = ord(text[i]) ^ ord(key[i % key_length])
        encrypted_bytes.append(encrypted_char)
    # Кодируем в base64 для удобства отображения
    return base64.b64encode(encrypted_bytes).decode('utf-8')

def xor_decrypt(encoded_text, key):
    encrypted_bytes = base64.b64decode(encoded_text)
    key_length = len(key)
    decrypted_text = ''
    for i in range(len(encrypted_bytes)):
        decrypted_char = encrypted_bytes[i] ^ ord(key[i % key_length])
        decrypted_text += chr(decrypted_char)
    return decrypted_text

# Шифр RSA
def rsa_generate_keys():
    (pubkey, privkey) = rsa.newkeys(2048)
    return pubkey, privkey

def rsa_encrypt(text, pubkey):
    encrypted = rsa.encrypt(text.encode('utf-8'), pubkey)
    return base64.b64encode(encrypted).decode('utf-8')

def rsa_decrypt(encoded_text, privkey):
    encrypted = base64.b64decode(encoded_text)
    decrypted = rsa.decrypt(encrypted, privkey)
    return decrypted.decode('utf-8')

# Шифр Плейфера
def playfair_generate_matrix(key):
    key = ''.join(sorted(set(key.upper()), key=lambda x: key.upper().index(x)))
    key = ''.join([c for c in key if c in alphabet])
    matrix = []
    used_letters = set()
    for c in key:
        if c not in used_letters:
            matrix.append(c)
            used_letters.add(c)
    for c in alphabet:
        if c not in used_letters:
            matrix.append(c)
            used_letters.add(c)
    extra_symbols = '0123456789'
    for c in extra_symbols:
        if len(matrix) >= 36:
            break
        matrix.append(c)
    return [matrix[i:i+6] for i in range(0, 36,6)]

def playfair_find_position(matrix, letter):
    for i, row in enumerate(matrix):
        for j, char in enumerate(row):
            if char == letter:
                return i, j
    return None, None

def playfair_encrypt(text, key):
    matrix = playfair_generate_matrix(key)
    text = ''.join([c for c in text.upper() if c in alphabet])
    i = 0
    new_text = ''
    while i < len(text):
        a = text[i]
        b = text[i+1] if i+1 < len(text) else 'Ъ'
        if a == b:
            new_text += a + 'Ъ'
            i += 1
        else:
            new_text += a + b
            i += 2
    if len(new_text) % 2 != 0:
        new_text += 'Ъ'
    result = ''
    for i in range(0, len(new_text), 2):
        a = new_text[i]
        b = new_text[i+1]
        row1, col1 = playfair_find_position(matrix, a)
        row2, col2 = playfair_find_position(matrix, b)
        if row1 == row2:
            result += matrix[row1][(col1+1)%6] + matrix[row2][(col2+1)%6]
        elif col1 == col2:
            result += matrix[(row1+1)%6][col1] + matrix[(row2+1)%6][col2]
        else:
            result += matrix[row1][col2] + matrix[row2][col1]
    return result

def playfair_decrypt(text, key):
    matrix = playfair_generate_matrix(key)
    result = ''
    for i in range(0, len(text), 2):
        a = text[i]
        b = text[i+1]
        row1, col1 = playfair_find_position(matrix, a)
        row2, col2 = playfair_find_position(matrix, b)
        if row1 == row2:
            result += matrix[row1][(col1-1)%6] + matrix[row2][(col2-1)%6]
        elif col1 == col2:
            result += matrix[(row1-1)%6][col1] + matrix[(row2-1)%6][col2]
        else:
            result += matrix[row1][col2] + matrix[row2][col1]
    result = result.replace('Ъ', '')
    return result

# Основная функция
def main():
    print("Выберите шифр:")
    print("1. Цезаря")
    print("2. Плейфера")
    print("3. XOR")
    print("4. RSA")
    print("5. Виженера")
    choice = input("Введите номер шифра: ")
    if choice == '1':
        text = input("Введите текст: ")
        key = int(input("Введите ключ (число): "))
        mode = input("Выберите режим (encrypt/decrypt): ")
        if mode == 'encrypt':
            print("Зашифрованный текст:", caesar_encrypt(text, key))
        elif mode == 'decrypt':
            print("Расшифрованный текст:", caesar_decrypt(text, key))
        else:
            print("Неверный режим")
    elif choice == '2':
        text = input("Введите текст: ")
        key = input("Введите ключевое слово: ")
        mode = input("Выберите режим (encrypt/decrypt): ")
        if mode == 'encrypt':
            print("Зашифрованный текст:", playfair_encrypt(text, key))
        elif mode == 'decrypt':
            print("Расшифрованный текст:", playfair_decrypt(text, key))
        else:
            print("Неверный режим")
    elif choice == '3':
        text = input("Введите текст: ")
        key = input("Введите ключ: ")
        mode = input("Выберите режим (encrypt/decrypt): ")
        if mode == 'encrypt':
            print("Зашифрованный текст:", xor_encrypt(text, key))
        elif mode == 'decrypt':
            print("Расшифрованный текст:", xor_decrypt(text, key))
        else:
            print("Неверный режим")
    elif choice == '4':
        text = input("Введите текст: ")
        mode = input("Выберите режим (encrypt/decrypt): ")
        if mode == 'encrypt':
            pubkey, privkey = rsa_generate_keys()
            encrypted_text = rsa_encrypt(text, pubkey)
            print("Зашифрованный текст:", encrypted_text)
            print("Приватный ключ (для расшифровки):", privkey.save_pkcs1().decode('utf-8'))
        elif mode == 'decrypt':
            encrypted_text = text
            privkey_data = input("Введите приватный ключ: ")
            privkey = rsa.PrivateKey.load_pkcs1(privkey_data.encode('utf-8'))
            print("Расшифрованный текст:", rsa_decrypt(encrypted_text, privkey))
        else:
            print("Неверный режим")
    elif choice == '5':
        text = input("Введите текст: ")
        key = input("Введите ключевое слово: ")
        mode = input("Выберите режим (encrypt/decrypt): ")
        if mode == 'encrypt':
            print("Зашифрованный текст:", vigenere_encrypt(text, key))
        elif mode == 'decrypt':
            print("Расшифрованный текст:", vigenere_decrypt(text, key))
        else:
            print("Неверный режим")
    else:
        print("Неверный выбор")

if __name__ == "__main__":
    main()
```
